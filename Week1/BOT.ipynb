```
def solve():
  n      = 0
  arr    = []
  p      = 0
  q      = 0
  profit = 0
  start  = 0

  n = int(input())
  arr = [int(x) for x in input().split(' ')]

  if n == 1:
    print(f'{0} {0} {arr[0]}')
    return

  for i in range(1, n):
    if arr[i-1] + arr[i] > arr[i]:
      arr[i] = arr[i-1] + arr[i]
    else:
      start = i

    if arr[i] > profit:
      profit = arr[i]
      p = start
      q = i

  print(f'{p+1} {q+1} {profit}')

solve()
```
